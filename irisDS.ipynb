```python
#Plotting the graph of sepal length and sepal width
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue="species", size=7) \
   .map(plt.scatter, 'sepal_length', 'sepal_width') \
   .add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(iris, hue='species', size=3)
plt.show()
```
